```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
```

```python
data_path = '../../../arquivos/car_prices.csv'
df = pd.read_csv(data_path).dropna()
df = df[df['year'] >= 2000]
df = df[['year', 'make', 'state','condition', 'color', 'mmr', 'sellingprice', 'odometer']]
marcas = df['make'].unique()
```

```python
def calcular_estatisticas(df, variavel):
    bli = {"marca": [], "media": [], "dam": [], "std": [], "mediana": [], "moda": [], "amplitude": []}
    
    for marca in marcas:
        dados_marca = df[df['make'] == marca][variavel]
        
        media = dados_marca.mean()
        dam = abs(media - dados_marca).mean()
        std = dados_marca.std(ddof=1)
        mediana = dados_marca.median()
        moda = dados_marca.mode().iloc[0]
        amplitude = dados_marca.max() - dados_marca.min()
        
        bli["marca"].append(marca)
        bli["media"].append(round(media, 2))
        bli["dam"].append(round(dam, 2))
        bli["std"].append(round(std, 2))
        bli["mediana"].append(mediana)
        bli["moda"].append(moda)
        bli["amplitude"].append(amplitude)
    
    return pd.DataFrame(bli)
```

```python
df_odometer_make = calcular_estatisticas(df, 'odometer')
df_condition_make = calcular_estatisticas(df, 'condition')
df_sellingprice_make = calcular_estatisticas(df, 'sellingprice')
df_mmr_make = calcular_estatisticas(df, 'mmr')
```

```python
display(df_odometer_make)
display(df_condition_make)
display(df_sellingprice_make)
display(df_mmr_make)
```
